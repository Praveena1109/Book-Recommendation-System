```python
book_titles = df1['BookTitle'].tolist()
```

```python
import urllib.request, urllib.parse, urllib.error
import json
import ssl

ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
serviceurl = 'https://www.googleapis.com/books/v1/volumes?'
# api_key = 'AIzaSyBGAxKXMaDHaX4Qbt7gVETB3Q08-Gs-dVM'
api_key = 'AIzaSyBUxiDRwhVfUc08_LE6HONcXFmV3520Pwc'
description = []
genre = []
previewlink = []
image = []
for x in book_titles:
    extensions = dict()
    extensions['q'] = x
    extensions['key'] = api_key
    url = serviceurl+urllib.parse.urlencode(extensions)
    try:
        urlh = urllib.request.urlopen(url, context=ctx)
        data = urlh.read().decode()
        js = json.loads(data)
    except:
        js = None
    try:
        d = js['items'][0]['volumeInfo']['description']
        description.append(d)
    except:
        description.append("NaN")
    try:
        g = js['items'][0]['volumeInfo']['categories'][0]
        genre.append(g)
    except:
        genre.append("NaN")
    try:
        t = js['items'][0]['volumeInfo']['imageLinks']['smallThumbnail']
        image.append(t)
    except:
        image.append("NaN")
    try:
        l = js['items'][0]['volumeInfo']['previewLink']
        previewlink.append(l)
    except:
        previewlink.append("NaN")
print(description)
print(genre)
print(previewlink)
print(image)

    
    
    
```

```python
df1['Description'] = description
df1['Genre'] = genre
df1['Previewlink'] = previewlink
df1['Image'] = image
```

```python
df1.AvgBookRating = df1.AvgBookRating.round(2)
```

```python
df1 = df1.drop_duplicates(subset ="BookTitle")
```
